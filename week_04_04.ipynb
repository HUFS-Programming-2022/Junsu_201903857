#### list comprehenshion
- 포함형/지능형/축약형 리스트
- 하나의 리스트에서 그를 포함한 또다른 리스트를 생성한다.
- "list + for loop"의 개념
- [A for B in C]
  - C에 있는 B에서 A를 가져와라 => 리스트에 담아서 반환
  - A: 사용자가 원하는 형태
  - B: C의 형태
  - C: 반복가능한 자료형 (리스트)
  
- 조건 추가: [A for B in C if ....]

### 리스트 <> 스트링
- split() : string.split(seperator)
    - default = 
- join()

```python
sent = 'I, have, a, car'
#['I', 'have', 'a', 'car'] # tokens
sent.split(',').strip() # strip()함수는 앞에 string이 있어야 사용 가능하기 때문에 오류가 생기는 경우
```

```python
#solution
sent_tokens = [word.strip() for word in sent.split(',')]
sent_tokens
```

```python
# ['I', 'have', 'a', 'car'] => 'I have a car'
'-'.join(sent_tokens)
```

### 반복문 맛보기
- 자료구조를 순회할 때 사용한다.
- 시작조건, 종료조건, 수행명령으로 구성
- for 변수 in <반복 범위>:
  <들여쓰기> <수행명령>
- 리스트 컴프리헨션의 [<수행명령> for 변수 in <반복 범위>] 구조와 흡사
- range(start, end, step): 숫자의 순회 시
    - 0부터 시작, 인덱스와 다른 점은 ->

```python
[word for word in sent_tokens]
```
