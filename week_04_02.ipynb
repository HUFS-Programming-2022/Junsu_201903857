#### 토큰: 프로그래밍에서 정의하는 단어
  - 단어, 혹은 기호 하나 하나를 토큰으로 볼 수 있다.

```python
len(moby.tokens)
```

```python
# 중복 단어를 걸러내기
set(moby.tokens)
```

- set
  - {a, b, c, ...}
  - 중복 불가능
  - 수학에서의 집합과 동일한 개념
  - 순서 없음
- list
  - [a, b, c, ...]
  - 중복 가능
  - 순서 있음

```python
# 중복 없는 단어의 수
len(set(moby.tokens))
```

```python
# lexical diversity
len(set(moby.tokens)) / len(moby.tokens)
```
